```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from sklearn.linear_model import LinearRegression
import numpy as np

# 字體設定（確保中文正常顯示）
font_path = "./font/NotoSansTC-VariableFont_wght.ttf"
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font_prop.get_name()
plt.rcParams['axes.unicode_minus'] = False

# 讀取資料並處理
df = pd.read_csv('investing_gold_2020_2025Feb.csv')
df.columns = df.columns.str.strip()
for col in ['Price', 'Open', 'High', 'Low']:
    df[col] = df[col].str.replace(',', '').astype(float)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# 計算每日漲跌
df['Price_Change'] = df['Price'].diff()

# 計算IQR與離群值界線
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 轉換日期為數值
df['Day'] = (df['Date'] - df['Date'].min()).dt.days

# 線性回歸模型
X = df[['Day']]
y = df['Price']
model = LinearRegression()
model.fit(X, y)

# 未來7天預測
future_days = np.arange(df['Day'].max()+1, df['Day'].max()+8).reshape(-1,1)
future_dates = pd.date_range(df['Date'].max() + pd.Timedelta(days=1), periods=7)
predictions = model.predict(future_days)

# --------------------------------------
# 🔁 黃金價格趨勢圖（隨時間變化）
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['Price'], color='#5bd1d7')
plt.title("黃金價格趨勢圖", fontproperties=font_prop, fontsize=16)
plt.xlabel("日期", fontproperties=font_prop)
plt.ylabel("價格（美元）", fontproperties=font_prop)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# --------------------------------------
# 📉 黃金價格預測：線性回歸圖（含歷史及未來預測）
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['Price'], label='歷史價格', color='#5bd1d7')
plt.plot(future_dates, predictions, label='未來7天預測', color='#d094ea', linestyle='--')
plt.title("黃金價格預測：線性回歸圖", fontproperties=font_prop, fontsize=16)
plt.xlabel("日期", fontproperties=font_prop)
plt.ylabel("價格（美元）", fontproperties=font_prop)
plt.legend(prop=font_prop)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# --------------------------------------
# 💡 離群值標示圖（在趨勢圖中標示異常）
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['Price'], label='價格', color='#5bd1d7')

# 標示離群值
outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]
plt.scatter(outliers['Date'], outliers['Price'], color='red', s=60, label='離群值', marker='o')

plt.title("黃金價格趨勢圖（離群值標示）", fontproperties=font_prop, fontsize=16)
plt.xlabel("日期", fontproperties=font_prop)
plt.ylabel("價格（美元）", fontproperties=font_prop)
plt.legend(prop=font_prop)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
