## Flask 是什麼？

Flask 是一個用 Python 語言開發的輕量級 Web 應用框架。它的設計理念是讓你可以快速且簡單地開始開發，同時又能夠隨著需求擴展成複雜的應用程式[^1][^2][^6][^8]。Flask 不會強制你使用特定的目錄結構或工具，而是讓開發者自由選擇所需的擴充功能，非常適合初學者學習 Web 開發[^2][^3][^9][^10]。

## Flask 的特色

- **輕量級**：核心功能精簡，初學者容易上手[^1][^2][^6][^8]。
- **彈性高**：不強制特定的資料庫、表單驗證或其他元件，開發者可依需求選擇擴充套件[^2][^6][^9][^10]。
- **易於擴展**：有大量社群提供的擴充功能（如資料庫、驗證、上傳檔案等）[^6][^7][^9]。
- **基於 WSGI 與 Jinja2**：WSGI 是 Python Web 應用的標準介面，Jinja2 則是強大的模板引擎，方便生成動態網頁[^7]。


## 基本結構與 Hello World 範例

以下是一個最簡單的 Flask 應用程式：

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello():
    return "Hello, World!"

if __name__ == "__main__":
    app.run()
```

- `from flask import Flask`：匯入 Flask 類別。
- `app = Flask(__name__)`：建立 Flask 應用實例。
- `@app.route("/")`：設定根目錄（/）的路由。
- `def hello():`：當訪問根目錄時，回傳 "Hello, World!"。
- `app.run()`：啟動本地開發伺服器，預設在 http://127.0.0.1:5000/[^6][^7][^8][^9]。


## 常用功能簡介

- **路由（Routing）**：使用 `@app.route()` 裝飾器定義 URL 與對應的處理函式[^7]。
- **模板渲染（Templates）**：透過 Jinja2 模板語法，讓 HTML 能動態顯示資料[^7]。
- **表單處理**：可接收並處理使用者提交的表單資料[^7]。
- **靜態檔案**：自動處理如 CSS、JavaScript、圖片等靜態資源[^7]。
- **擴充功能**：如 Flask-SQLAlchemy（資料庫）、Flask-WTF（表單）、Flask-Login（登入認證）等[^7][^10]。


## 適合何時使用 Flask？

- 適合學習 Web 開發、快速原型設計、小型或中型專案[^2][^9]。
- 若專案規模很大或需要高度整合的功能，可以考慮 Django 等更完整的框架[^9]。


## 小結

Flask 是 Python 生態系中最受歡迎且適合初學者的 Web 框架之一。它簡單、彈性高、學習曲線平緩，非常適合想用 Python 開發網站或 API 的新手[^2][^3][^6][^7][^9][^10]。

Dependencies
These distributions will be installed automatically when installing Flask.

Werkzeug implements WSGI, the standard Python interface between applications and servers.

Jinja is a template language that renders the pages your application serves.

MarkupSafe comes with Jinja. It escapes untrusted input when rendering templates to avoid injection attacks.

ItsDangerous securely signs data to ensure its integrity. This is used to protect Flask’s session cookie.

Click is a framework for writing command line applications. It provides the flask command and allows adding custom management commands.

Blinker provides support for Signals.
